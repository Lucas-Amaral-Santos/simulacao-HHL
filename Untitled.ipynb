```python
from qiskit.tools.jupyter import *
from qiskit import QuantumCircuit, IBMQ, BasicAer
from qiskit.extensions import UnitaryGate
from qiskit.visualization import plot_histogram
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

import qiskit


from qiskit_ibm_provider import IBMProvider, least_busy
from qiskit_ibm_runtime import QiskitRuntimeService



import qiskit.quantum_info as qi

from math import cos,sin, pi, e, asin
```

```python
def make_u(matrix,i=0):
    u = qi.Operator(matrix)
    
    gate = UnitaryGate(u, label=f"U{2**i}")
    return gate
```

```python
def calcula_t(lamb, lamb_cod, N):
    return ((lamb_cod*2*pi)/((2**N)*lamb))
```

```python
U_test = []
U_test2 = []

def hhl_2x2(qc, n, A):

    w,v = np.linalg.eig(A)

    print("V: "+ str(v))
    v[0][1] = -v[0][1]
    v[1][1] = -v[1][1]

    # need to be binary represented with n-2 bits 
    lambda1 = round(min(w)/min(w))
    lambda2 = round(max(w)/min(w))
    
    print("Lambda1: " + str(min(w)))
    print("Lambda2: " + str(max(w)))

    print("Lambda1 codificada: " + str(lambda1))
    print("Lambda2 codificada: " + str(lambda2))

    t = calcula_t(min(w), lambda1, n-2)
    
    print("t: ", t)
    print("Gabriela exp(i*2*pi*lambda*N): " + str(e**(1j*(2*pi*lambda1/2**(n-2)))))
    print("Lucas exp(i*lambda*t): " + str(e**(1j*min(w)*t)))

    U_array = [] # in array format to be retrieved later
    

    for i in range(n-2):
        if i == 0:
            # Udiag = [[e**(1j*(2*pi*lambda1/2**(n-2))),0],[0,e**(1j*(2*pi*lambda2/2**(n-2)))]]
            Udiag = [[e**(1j*min(w)*t),0], [0, e**(1j*max(w)*t)]]
            U_test2.append(Udiag)
            #Udiag = [[1,0],[0, 1]]

        else:
            Udiag = np.dot(Udiag,Udiag)
            U_test2.append(Udiag)
        U_array.append(np.dot(np.dot(v,Udiag),v.transpose().conjugate()))


    U_test.append(U_array)
    qc.barrier()
    for i in range(1, n-1):
        qc.h(i)
    qc.barrier()

    state1 = qiskit.quantum_info.Statevector(qc)
    
    for i in range(1, n-1):
        qc.append(make_u(U_array[i-1],i-1).control(1),[i, n-1])
    qc.barrier()

    state2 = qiskit.quantum_info.Statevector(qc)

    qc.append(qiskit.circuit.library.QFT(n-2,inverse=True),range(1, n-1))
    qc.barrier()

    state3 = qiskit.quantum_info.Statevector(qc)

    # for i in range(1, n-1):
    #     qc.measure(i,i-1)

    qc.append(qiskit.circuit.library.RYGate(2*asin(1/lambda1)).control(n-2, ctrl_state=bin(lambda1)[2:].zfill(n-2)),[i%(n-1) for i in range(1,n)])
    qc.append(qiskit.circuit.library.RYGate(2*asin(1/lambda2)).control(n-2, ctrl_state=bin(lambda2)[2:].zfill(n-2)),[i%(n-1) for i in range(1,n)])
    qc.barrier()
    state4 = qiskit.quantum_info.Statevector(qc)

    qc.append(qiskit.circuit.library.QFT(n-2,inverse=False),range(1, n-1))
    qc.barrier()

    state5 = qiskit.quantum_info.Statevector(qc)

    for i in range(n-2, 0, -1):
        qc.append(make_u(np.linalg.inv(U_array[i-1]),i-1).control(1),[i, n-1])
    qc.barrier()    

    for i in range(1, n-1):
        qc.h(i)
    qc.barrier()

    # Measure
    state = qiskit.quantum_info.Statevector(qc)
    prob = state.probabilities()


    qc.measure(0,0)
    qc.measure(n-1,1)

    return state1, state2, state3, state4, state5, prob, qc
```

```python
A = [[0.0089, 0.0116], [0.0116, 0.0188]]# [[6, 4], [4, 6]] # [[0.0089, 0.0116], [0.0116, 0.0188]]
n = 7
qc = QuantumCircuit(n,2)

#qc.x(n-1)
# qc.h(n-1)

e1, e2, e3, e4, e5, prob,qc = hhl_2x2(qc, n, A) 

qc.draw('mpl')
```
